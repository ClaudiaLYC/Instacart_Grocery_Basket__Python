### If-Statements with User-Defined Functions

#### Create a ´Price_label´: 3 different prices ranges

```python
# Defining a Price_label function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return 'Not enough data'
```

```python
# Apply the new function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Check what is the most expensive product 
df['prices'].max()
```

## Task 4.7 Deriving New Variables

### Step 2: New Column for two busiest and two slowest days of the week

```python
# First, find the busiest and slowest days from the ´orders_day_of_the_week`column
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

#### The two busiest days are Saturday (0) and Sunday (1)
#### The two slowest days are Wednesday (4) and Tuesday (3)

```python
# Second: define the for-loop and create an empty list called ´result_days´

result_days = []

for value in df_ords_prods_merged['orders_day_of_week']:
    if value == 0 or value == 1:
        result_days.append('Busiest days')
    elif value == 4 or value == 3:
        result_days.append('Slowest days')
    else:
        result_days.append('Normal days')
```

```python
# Add result_days to dataframe in a new column
df_ords_prods_merged['Busiest_days'] = result_days
```

#### Step 4: Create a new column containing labels `Busiest_period_of_day´

```python
# Check the required column on which to focus first
df_ords_prods_merged.head()
```

```python
# Check the frequency of values in the "order_hour_of_day" column 
df_ords_prods_merged['order_hour_of_day'].value_counts()
```

#### The request is to have 3 periods of time, hence we need to divide the 24 entries (ie hours) into 3 labels of 8 values as such: “Most orders” (10,11,14,15,13,12,16,9), “Average orders” (17,8,18,19,20,7,21,22), and “Fewest orders” (23,6,0,1,5,2,4,3) 

```python
# Create an If-statement with For-loop using the above 3 requested labels

time_of_orders = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        time_of_orders.append('Most Orders')
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        time_of_orders.append('Fewest Orders')
    else:
        time_of_orders.append('Average Orders')
```

```python
# Create a new column "Busiest_period_of_day" with the result printed in it

df_ords_prods_merged['busiest_period_of_day'] = time_of_orders
```
