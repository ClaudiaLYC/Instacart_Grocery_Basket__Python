## Subsetting

```python
df_prods
```

```python
# Create a subset for df_prods that only contains data from the snacks department (of value =19)
df_snacks = df_prods[df_prods['department_id']==19]
```

```python
df_snacks.head()
```

```python
# Method 2 for creating a subset as above using loc
df_snacks_2 = df_prods.loc[df_prods['department_id']  == 19]
```

```python
df_snacks_2.head()
```

```python
# Method 3 using isin to locate more than one value
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([18,19])]
```
