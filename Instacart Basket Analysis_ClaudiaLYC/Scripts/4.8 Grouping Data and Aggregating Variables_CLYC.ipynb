### Grouping Data with pandas

#### First: Split the data into groups based on the given criteria
#### groupby() function: will group a given dataframe by a given column

```python
# Use the groupby function first as example on the product_name

df.groupby('product_name')
```

#### Second: Aggregating  Data with agg()

```python
# Calculate the mean of the ´order_number` column grouped by the department_id column to compare the average nummber of orders per user
# 1. Split the data into groups on department_id

df.groupby('department_id').agg({'number_of_orders': ['mean']})
```

```python
# Aggregation conducted withouth the agg() function  - same result as above but different format

df.groupby('department_id')['number_of_orders'].mean()
```

#### Performing Multiple Aggregations

```python
# To produce multiple statistics at the same time, such as mean, minimum, maximum

df.groupby('department_id').agg({'number_of_orders': ['mean', 'min', 'max']})
```

#### Third: Aggregrating Data with Transform() - creating flags for the data

```python
# Create max_order column: three steps in one line of code
# 1. Split data into groups based on user_id, 2. Apply transform() on order_number to generate max orders of each user, 3. create max_order for the results 

df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['number_of_orders'].transform(np.max)
```

#### Deriving Columns with loc()

##### Create the flag with following criteria:

##### 1. If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
##### 2. If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
##### 3. If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

```python
# Condition #1 above
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
# Condition #2 above
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
# Condition #3 above
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

#### Step 6: Create a spending flag for each user based on the average price across all their orders using the following criteria:
#### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
#### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Create a new column ´average_price´ and aggregate with transform() using numpy mean

df_ords_prods_merge['average_price']=df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check output
df_ords_prods_merge.head(25)
```

```python
# Create ´Spending Flag´ column and assign labels ´Low spender´ for prices under 10

df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Create ´Spending Flag´ column and assign labels ´High spender´ for prices equal or above 10

df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check output

df_ords_prods_merge.head()
```

```python
# Check the frequency of new column ´spending flag´

df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

#### Step 7: Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create a new column ´median_prior_order´ and aggregate with transform() using numpy median

df_ords_prods_merge['median_prior_order']=df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
# Check output

df_ords_prods_merge.head(25)
```

```python
# Create new column ´freqency flag´ based on median days since prior order and label ´Non-frequent Customer´ when higher than 20
df_ords_prods_merge.loc[df_ords_prods_merge['median_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Create new column ´freqency flag´ based on median days since prior order and label ´Regular Customer´ when higher than 10 and lower or equal 20
df_ords_prods_merge.loc[(df_ords_prods_merge['median_prior_order'] > 10) & (df_ords_prods_merge['median_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# Create new column ´freqency flag´ based on median days since prior order and label ´Frequent Customer´ is lower than or equal to 10
df_ords_prods_merge.loc[df_ords_prods_merge['median_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
